##Cleaning and Preprocessing data

```python
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies["keywords"].iloc[0]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
credits.head()
```

```python
credits['cast'].iloc[0]
```

```python
import ast
def extract(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt != 3:
            L.append(i['name'])
            cnt += 1
        else:
            break
    return L
```

```python
credits['cast'] = credits['cast'].apply(extract)
credits.head()
```

```python
credits['crew'].iloc[0]
```

```python
import ast

def pull_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
credits['crew'] = credits['crew'].apply(pull_director)
credits.head()
```

```python
credits.rename(columns={'movie_id': 'id'}, inplace=True)
```

```python
new_df = movies.merge(credits, on='id')
new_df.head()
```

```python
import pandas as pd

# Handle potential float values in the 'overview' column
new_df['overview'] = new_df['overview'].apply(lambda x: x.split() if isinstance(x, str) else [])
new_df.head()
```

```python
new_df['genres'] = new_df['genres'].apply(lambda x:[i.replace(" ","")for i in x])
new_df['keywords'] = new_df['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
new_df['cast'] = new_df['cast'].apply(lambda x:[i.replace(" ","")for i in x])
new_df['crew'] = new_df['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['overview'] + new_df['genres'] + new_df['keywords'] + new_df['cast'] + new_df['crew']
```

```python
new_df.head()
```

```python
new_df = new_df[['id', 'title', 'tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

## Main function

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
