#**Agricultural Production Analysis of Bihar,India**

```python
# Importing Required Package

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#####*The Dataset Contains Data about Agricultural Production and Their Parameter for all Indian States From 1997-2020*

```python
# Opening the Dataset
dataframe = pd.read_csv('/content/drive/MyDrive/python pandas/crop_yield.csv')
```

#####**Extraction of Data Related to Bihar State from Dataset containing Data of All states from 1997-2019**

```python
# Extracting Agricultural Production Data Specific to Bihar State
Bihar_data = dataframe[dataframe['State'] == 'Bihar']
Bihar_data.head(-5)
```

####**General Statical Discription of Dataset**

```python
dataframe.describe()
```

###**Analysis--**

###**Analysis(1)**
#####**Crop Vs Production in Bihar Using Scatter Plotting**

```python
# Normalisation of Production Using Simple Feature Scaling Normalisation Method

data_production = np.array(Bihar_data['Production'])
max = data_production.max()
data_Nproduction = data_production/max
```

```python
# Scatter Plotting Crop V/S Production

plt.scatter(data_Nproduction,Bihar_data['Crop'],s=15)
plt.title('Crop vs Production 1997-2019')
plt.xlabel('Production')
plt.ylabel('Crop')
```

### Analysis Result:
**Sugarcane** is the highest production crop in state of Bihar . In terms of Production.

Other Major crop is Rice



### **Analysis(2):**

###**Crop Vs Yield in Bihar Using Scatter Plotting**

```python
''' Normalisation of Yield Using Simple Feature Scaling Normalisation Method '''

data_yield = np.array(Bihar_data['Yield'])
max = data_yield.max()
data_Nyield = data_yield/max
```

```python
# Scatter Plotting Crop V/S Yield

plt.scatter(data_Nyield,Bihar_data['Crop'],s=15)
plt.title('Crop vs Yield 1997-2019')
plt.xlabel('Yield')
plt.ylabel('Crop')
```

### Analysis Result:
**Sugarcane** is the crop having highest Yield in State of Bihar.

Other crop having high yield is **Banana** followed by **Jute** and **Mesta**






###**Analysis(3)**

###Production over the year in Bihar Using Scatter Plots

```python
# Grouping of Data Across Year using SUM
# To Obatin The net Production in an Year
# i.e Sum of production of all Crops in a year

data_YearProduction = Bihar_data[['Crop_Year','Production']]
data_netproduction = data_YearProduction.groupby(['Crop_Year']).sum()
years = Bihar_data['Crop_Year'].unique()

# Normalising Net Production
# Using Simple Feature Scaling Method

data_netproduction = np.array(data_netproduction)
max = data_netproduction.max()
data_Nnetproduction = data_netproduction/max
```

```python
# Scatter Plotting Year vs Production

plt.scatter(years,data_Nnetproduction,s=15)
plt.title('Year vs Net Production 1997-2019')
plt.xlabel('Year')
plt.ylabel('Production')
plt.xticks(np.arange(1997,2021,2))
plt.grid(True)
```

###**Analysis Result:**
The ***net agricultural production*** of Bihar is highest in year **2018**. Year **2006** was a year of sudden dip in agriculture production

###**Analysis(4):**

**Fertilizer v/s Production and Yield in Bihar using Scatter Plots**

```python
# Scatter Plotting

# Fertilizer vs Production
plt.subplot(2,1,1)
plt.scatter(Bihar_data['Fertilizer'],Bihar_data['Production'],s=15)
plt.title('Fertilizer vs Production')
plt.xlabel('Fertilizer')
plt.ylabel('Production')

# Fertilizer vs Yield
plt.subplot(2,1,2)
plt.scatter(Bihar_data['Fertilizer'],Bihar_data['Yield'],s=15)
plt.title('Fertilizer vs Yield')
plt.xlabel('Fertilizer')
plt.ylabel('Yield')

# Formatting subplots
plt.subplots_adjust(hspace=0.7)
```

###**Analysis Result**
In the first plot as the Fertilizer increases then the Production also Increases.

In second Plot as the Fertilizer increases then Yield is Decreases

**Pesticides v/s Production and Yield in Bihar using Scatter Plots**

```python
# Scatter Plotting

# Pesticides vs Production
plt.subplot(2,1,1)
plt.scatter(Bihar_data['Pesticide'],Bihar_data['Production'],s=15)
plt.title('Pesticides vs Production')
plt.xlabel('Pesticides')
plt.ylabel('Production')

# Pesticides vs Yield
plt.subplot(2,1,2)
plt.scatter(Bihar_data['Pesticide'],Bihar_data['Yield'],s=15)
plt.title('Pesticides vs Yield')
plt.ylabel('Yield')
plt.xlabel('Pesticides')

# Formatting subplots
plt.subplots_adjust(hspace=0.7)
```

###**Analysis(6):**

**Annual Rainfall v/s Production and Yield in Bihar using Scatter Plots**

```python
# Scatter Plotting

# Rainfall vs Production
plt.subplot(2,1,1)
plt.scatter(Bihar_data['Annual_Rainfall'],Bihar_data['Production'],s=15)
plt.title('Annual Rainfall vs Production')
plt.xlabel('Rainfall')
plt.ylabel('Production')

# Rainfall vs Yield
plt.subplot(2,1,2)
plt.scatter(Bihar_data['Annual_Rainfall'],Bihar_data['Yield'],s=15)
plt.title('Annual Rainfall vs Yield')
plt.ylabel('Yield')
plt.xlabel('Rainfall')

# Formatting subplots
plt.subplots_adjust(hspace=0.7)
```
